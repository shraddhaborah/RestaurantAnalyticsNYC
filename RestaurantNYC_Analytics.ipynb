```python

df_subset = df[[
    'Restaurant Name', 'Doing Business As (DBA)', 'Street',
    'Borough', 'Postcode', 'Time of Submission',
    'Latitude', 'Longitude', 'NTA', 'Seating Interest (Sidewalk/Roadway/Both)'
]].copy()


df_subset['Time of Submission'] = pd.to_datetime(
    df_subset['Time of Submission'],
    format='%m/%d/%Y %I:%M:%S %p',
    errors='coerce'
)
```

```python
# Filter restaurants submitted in the last 60 days
recent_threshold = pd.Timestamp.today() - pd.Timedelta(days=60)
recent_restaurants = df_subset[df_subset['Time of Submission'] > recent_threshold]

# Group by borough or NTA
recent_count = recent_restaurants.groupby('Borough').size().sort_values(ascending=False)
print("New Restaurants by Borough:\n", recent_count)

# Sample output for dashboard
recent_restaurants[['Restaurant Name', 'Street', 'Borough', 'Time of Submission']].head()
```

```python


# Load data
df = pd.read_csv("cleaned_restaurants.csv")
df['Time of Submission'] = pd.to_datetime(df['Time of Submission'])
```

```python
import streamlit as st
st.title("NYC Emerging Restaurant Insights")
```

```python
# Filter for recent
recent = df[df['Time of Submission'] > pd.Timestamp.today() - pd.Timedelta(days=60)]

# Borough filter
boroughs = st.multiselect("Select Borough(s):", df['Borough'].unique(), default=df['Borough'].unique())
filtered = recent[recent['Borough'].isin(boroughs)]
```

```python
import plotly.express as px
st.subheader("🗺️ Restaurant Map")
fig = px.scatter_mapbox(
    filtered,
    lat='Latitude',
    lon='Longitude',
    hover_name='Restaurant Name',
    color='Borough',
    zoom=10,
    mapbox_style="open-street-map"
)
st.plotly_chart(fig)
```

```python
st.subheader("📊 Seating Type Distribution")
st.bar_chart(filtered['Seating Interest (Sidewalk/Roadway/Both)'].value_counts())
```
